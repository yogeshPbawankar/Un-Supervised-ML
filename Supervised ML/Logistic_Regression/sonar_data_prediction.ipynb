```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading dataset to a pandas dataframe
sonar_data = pd.read_csv(r'C:\Users\Lenovo\Desktop\myenv\Data\sonar.csv')
```

```python
sonar_data.head()
```

```python
# number of rows and column 
sonar_data.shape
```

```python
# statistical measure of the data
sonar_data.describe()
```

```python
# how many example is there specify the col
sonar_data.value_counts()
```

```python
# M --> Mine 
# R --> Rock
X = sonar_data.drop(columns = 'R',axis=1)
Y = sonar_data['R']
```

```python
print(sonar_data.columns)
```

```python
print(X)
print("_____________________________________________________________________")
print(Y)
```

# Training and Testing Data

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=1)
```

```python
print(X.shape,x_train.shape,x_test.shape)
```

# Model Training

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

# Model Evalutation

```python
# Accuracy on the training data
x_train_prediction = model.predict(x_train)
traning_data_accuracy = accuracy_score(x_train_prediction,y_train)
print("accuracy_score on training data : ",traning_data_accuracy)
```

```python
# Accuracy on the testing data
x_test_prediction = model.predict(x_test)
testing_data_accuracy = accuracy_score(x_test_prediction,y_test)
print("accuracy_score on test data : ",testing_data_accuracy)
```
